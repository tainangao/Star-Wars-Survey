# Import data

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head(10)
```

```python
star_wars.columns
```

# Rename columns

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6",
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6",
        'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.':
        'Unnamed: 15',
        "Have you seen any of the 6 films in the Star Wars franchise?":'seen_star_wars',
        "Do you consider yourself to be a fan of the Star Wars film franchise?":'fan_star_wars',
        'Are you familiar with the Expanded Universe?':'familiar_expanded_universe',
        'Do you consider yourself to be a fan of the Expanded Universe?ÂÃ¦':'fan_expanded_universe',
        'Do you consider yourself to be a fan of the Star Trek franchise?':'fan_star_trek',
        'Location (Census Region)':'location'
        })

star_wars.columns
```

```python
star_wars.columns = star_wars.columns.str.lower().str.replace(': ','_').str.replace(' ','_')
star_wars.columns
```

# Set boolean values

```python
yes_no = {"Yes": True, "No": False}
for col in [
    "seen_star_wars",
    "fan_star_wars",
    'familiar_expanded_universe',
    'fan_expanded_universe',
    'fan_star_trek'
    ]:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

# Null values

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


def plot_null_matrix(df, figsize=(15,13)):
    # initiate the figure
    plt.figure(figsize=figsize)
    # create a boolean dataframe based on whether values are null
    df_null = df.isnull()
    # create a heatmap of the boolean dataframe
    sns.heatmap(~df_null, cbar=False, yticklabels=False)
    plt.xticks(rotation=90, size='x-large')
    plt.show()
    
    
plot_null_matrix(star_wars)

# black indicates null value
```

```python
# drop rows with null values
star_wars = star_wars[star_wars['fan_star_wars'].notnull()]


# drop the whole column
star_wars = star_wars.drop('fan_expanded_universe', 1)

plot_null_matrix(star_wars)
```

# Rankings
<b>In the original dataset, 1 means the film was the most favorite, and 6 means it was the least favorite. When we plot this on a bar plot, it's misleading. To make it more intuitive and Coherent with the previous plots, I'll reverse the values.</b>

### Reverse coding


```python
rank_cols = star_wars.columns[9:15].values
re_rank = {"1": '6', "2": '5', '3':'4', '4':'3' ,'5':'2', '6':'1' }

for col in rank_cols:
    star_wars[col] = star_wars[col].map(re_rank)
```

### Convert data type

```python
star_wars[rank_cols] = star_wars[rank_cols].astype(float)

star_wars[rank_cols].mean().plot()
```

Just like the # of view, rank_5 (Star Wars: Episode V The Empire Strikes Back) received the best rank and rank_3 (Star Wars: Episode III Revenge of the Sith) received the worst rank.

```python
fig = plt.figure(figsize=(10,18))

for sp in range(0,5):
    ax = fig.add_subplot(5,1,sp+1)
    table = pd.pivot_table(star_wars,index=cat_cols[sp],values=rank_cols, fill_value=0, aggfunc=np.mean)
    ax = sns.heatmap(table, cmap='GnBu', annot=True,cbar=False)
    ax.set(ylabel='')
    ax.set_title(cat_cols[sp],fontsize=20),
    
    
plt.subplots_adjust(hspace = 0.5)    
plt.show()
```

Generally speaking, movies have more views also receiver better ranks.

- gender
    - contrary to views, male and female don't seem to rank differently
- age
    - 60+ years old respondents enjoy ranking_1 (Star Wars: Episode I  The Phantom Menace) a lot more than other age groups
    - 60+ years old respondents enjoy the last 3 movies less than other age groups
    - 30-44 years old respondents enjoy the last 3 movies more than other age groups
- household_income
    - ranking_4 (Star Wars: Episode IV  A New Hope) is the most controversial movie
- education
    - respondents who don't have highshcool degree seem to give more extreme rankings

# Characters

#### It appears that these unnamed columns are characters' names

```python
char_cols = star_wars.columns[15:29].values
star_wars[char_cols].head()
```

```python
fig = plt.figure(figsize=(10,30))

for sp in range(0,14):
    ax = fig.add_subplot(14,1,sp+1)
    ax = sns.countplot(data=star_wars, y=char_cols[sp], palette="Spectral",
                       order=star_wars['unnamed_15'].value_counts().index)
    ax.set(ylabel='',xlabel='')
    ax.set_title(char_cols[sp],fontsize=15),
    ax.set_xlim(0, 650)
    ax.ylim=(0,5)

    
    
plt.subplots_adjust(hspace = 0.8)    
plt.show()
```

The most controversial characters are unnamed_20, unnamed_21, unnamed_26.

```python
fav = {"Unfamiliar (N/A)": 0,"Very unfavorably": 1,  'Somewhat unfavorably':2, 
       'Neither favorably nor unfavorably (neutral)':3 ,'Somewhat favorably':4, 'Very favorably':5 }

characters = star_wars[char_cols].copy()
for col in char_cols:
    characters[col] = characters[col].map(fav)
    
characters.head()
```

```python
characters.mean()
```

```python
fig, ax = plt.subplots(figsize=(7, 6))
ax.bar(x = np.arange(14), height = characters.mean())
ax.set_title('Character Popularities', size='x-large')
ax.set_ylabel('Average Points', size='x-large')
ax.set_xticks(np.arange(14))
ax.set_xticklabels(char_cols, rotation=90, size='x-large')

fig.tight_layout()
```

The least popular characters are unnamed_20 and unnamed_26.
